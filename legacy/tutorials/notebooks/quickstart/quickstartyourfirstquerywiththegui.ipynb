# Quick Start - Your First Query with the Graphical User Interface
## Geospatial Analytics Credentials
To access the user interface of the [Geospatial Analytics](https://ibmpairs.mybluemix.net) component of the [IBM Environmental Intelligence Suite](https://www.ibm.com/products/environmental-intelligence-suite?utm_medium=OSocial&utm_source=Linkedin&utm_content=DTPWW&utm_id=EIS-MM-ProductPage&social_post=5712196498&linkId=135589025) you will need a username and password. The username will typically be your IBMID but if you don't have credentials then please contact your IBM representative.

## Logging In

Navigate to the [IBM Environment Intelligence Suite login page](https://ibmpairs.mybluemix.net) and enter your credentials. This should take you to a [Queries](https://ibmpairs.mybluemix.net/queries) page (see below). If this is the first time you have used Geospatial Analytics, your page will have no 'Recent Queries' displayed. 

![Queries1](guiqueries1.jpg)

If you press the "open menu" button in the very top left hand corner of the screen (circled in the next screenshot) then you can see how to navigate around the Environmental Intelligence Suite:

![Queries Menu](guiqueriesmenu.jpg)

You should see submenus for the three default components:

 * Dashboard Visualizations
 * Geospatial Analytics
 * Alert Console

For now we're going to get you started by running a simple query in Geospatial Analytics using the "Queries" page. 

## Creating A Point Query
We're going to create a point query, so press the "Create Query" button in the top right hand corner (circled in the previous screenshot).

This leads to a series of screens where you specify what data you want to query, where in the world you want to query and the time frame you are interested in. For this first query:

 * What = Temperature above ground from the Current and historical weather (IBM TWC) data set
 * Where = A point in the United Kingdom (UK)
 * When = 12 noon (UTC) on 1st December 2021

### What (data layers you want to query)

Press the "add Layers" button and type "IBM TWC" into the search box (circled in the following screen shot). 

![Add Layers](guiaddlayers.jpg)

This will produce a list of the data layers in the "Current and historical weather (IBM TWC)" data set. This is the weather data produced with The Weather Company. If you scroll through the list of data layers you should find "Temperature above ground". Select the row so that the tick symbol appears (circled) and then press "Add". This will take you back to the "what" page but with the selected layer included. 

![What](guiwhat.jpg)

Press the "Next" button.

### Where (in the world are you interested in seeing data for)
On the "Where" page select the "Point" button (circled in the following screenshot), move and size the map so that you see the UK and then click to put a point somewhere. 

![Where](guiwhere.jpg)

Press the "Next" button.

### When (in time are you interested in seeing data for)
On the "When" page set the "From" and "To" data/time field for the data you are interested in. In this case I have set them to the hour each side of 12 noon on the 1st December 2021. 

![Where](guiwhen.jpg)

Press the "Next" button.

## Running The Query
The "Run" page shows a summary of the query options you have specified and lets you give it a name (circled in the screenshot below). 

![Run](guirun.jpg)

Press the "Run" button.

## Looking At the Results
You will be directed back to the "Queries" page and, after a short while, your new query should appear in the top left hand corner (circled in the screenshot below).

![Queries2](guiqueries2.jpg)

Select the query's tile and the results page will open.

![Results1](guiresults1.jpg)

In the panel to the left you will see information about the query you ran, its name, the layers and the time range you selected. If you click a point on the map you will see the data in the time-series display at the bottom  change (outlined in the screenshot above). The time-series display shows you the selected data layer value (temperature above ground - in Kelvin) at hourly intervals (the interval at which the data is generated) within the selected time range at the point you selected on the map.

The point you selected when creating the query is just the first point it displays and you can click anywhere on the map and get the temperature values. 

The date and time range you selected when creating the query is just the fist time range that it displays and you can chose any time range in the time range display (circled in the screenshot below).

![Results2](guiresults2.jpg)

As we extend the range of the time-series being displayed the seasonal variation in temperature becomes evident. 

You can download this timeseries data to your computer by pressing the "Download as csv" button (circled).
